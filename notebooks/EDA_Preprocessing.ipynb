# Pre-processing

```python
print("Missing Values:\n")
print(df.isnull().sum())
```

```python
# DateTime feature extraction

df['timestamp'] = pd.to_datetime(df['timestamp'])

# Extract datetime features
df['hour'] = df['timestamp'].dt.hour
df['day'] = df['timestamp'].dt.day
df['month'] = df['timestamp'].dt.month
df['weekday'] = df['timestamp'].dt.weekday
df.drop(columns=['timestamp'], inplace=True)
```

```python
# Identify numerical and categorical columns

numeric_cols = df.select_dtypes(include=[np.number]).columns
cat_cols = df.select_dtypes(include='object').columns
```

```python
# Impute missing values with median

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
df[numeric_cols] = imputer.fit_transform(df[numeric_cols])
```

```python
print("After imputation missing values:\n")
print(df.isnull().sum())
```

```python
df_before = df.copy()
```

# Outlier

```python
for col in numeric_cols:
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

    sns.boxplot(x=df_before[col], ax=axes[0], color='orange')
    axes[0].set_title(f'Before Outlier Removal - {col}')
    
    plt.tight_layout()
    plt.show()
```

```python
# Outlier removal usin IQR method

Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
for col in numeric_cols:
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

    sns.boxplot(x=df[col], ax=axes[1], color='skyblue')
    axes[1].set_title(f'After Outlier Removal - {col}')
    
    plt.tight_layout()
    plt.show()
```

```python
# Recompute numeric cols after dropping outliers

numeric_cols = df.select_dtypes(include=[np.number]).columns
print(numeric_cols)
```

```python
# Feature scaling using Standard Scaler

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
```

# Multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

X_vif = df[numeric_cols]
vif_data = pd.DataFrame()
vif_data["feature"] = X_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(len(X_vif.columns))]

print("\n📊 VIF Scores:\n", vif_data)
```

```python
# Drop irrelevant features

irrelevant_features = [
    'timestamp',  
    'atmospheric_pressure',
    'wind_speed',
    'visibility_index',
    'dew_point',
    'random_variable1',
    'random_variable2',
    
]

df.drop(columns=[col for col in irrelevant_features if col in df.columns], inplace=True)
```

```python
# Select only numeric columns for VIF calculation

X = df.select_dtypes(include=[np.number])

def calculate_vif(X):
    vif_data = pd.DataFrame()
    vif_data["feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data


threshold = 10
while True:
    vif_df = calculate_vif(X)
    high_vif = vif_df[vif_df["VIF"] > threshold]
    if high_vif.empty:
        break
    col_to_drop = high_vif.sort_values("VIF", ascending=False).iloc[0]["feature"]
    print(f"Dropping '{col_to_drop}' with VIF = {high_vif['VIF'].max():.2f}")
    X = X.drop(columns=[col_to_drop])


df = df[X.columns]

print("\nRemaining columns after VIF filtering:")
print(df.columns.tolist())
print(f"\nFinal Dataset Shape: {df.shape}")
```
